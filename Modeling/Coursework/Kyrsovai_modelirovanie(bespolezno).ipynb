```python
b1 = 0.5
b2 = 0.25
b3 = 0.5
b4 = 0.5

a1 = 2
a2 = 0.5
a3 = 0.2
a4 = 0.3

k1 = 2
k2 = 1
k3 = 1
k4 = 2
```

# Для ЗСеМО

```python
def m(M, i):
  if i == 1:
    return a1 * u(M, i) * proizv(M)
  if i == 2:
    return a2 * u(M, i) * proizv(M)
  if i == 3:
    return a3 * u(M, i) * proizv(M)
  if i == 4:
    return a4 * u(M, i) * proizv(M)

def u(M, i):
  wtf = 1 + (0 if M == 1 else m(M-1, i))
  if i == 1:
    return b1 * wtf
  if i == 2:
    return b2 * wtf
  if i == 3:
    return b3 * wtf
  if i == 4:
    return b4 * wtf
  # if i == 1:
  #   return b1 * wtf if M != k1 else b1
  # if i == 2:
  #   return b2 * wtf if M != k2 else b2
  # if i == 3:
  #   return b3 * wtf if M != k3 else b3
  # if i == 4:
  #   return b4 * wtf if M != k4 else b4
  

def U(M):
  return a1 * u(M, 1) + a2 * u(M, 2) + a3 * u(M, 3) + a4 * u(M, 4)

def proizv(M):
  return M / (U(M))

def _w(M, i):
  if i == 1:
    return u(M, 1) - b1
  if i == 2:
    return u(M, 2) - b2
  if i == 3:
    return u(M, 3) - b3
  if i == 4:
    return u(M, 4) - b4

def W(M):
  return a1 * (u(M, 1) - b1) + a2 * (u(M, 2) - b2) + a3 * (u(M, 3) - b3) + a4 * (u(M, 4) - b4)

def L(M):
  return W(M) * proizv(M)
```

# Для РСеМО

```python
def proizv_i(proizv, i):
  if i == 1:
    return a1 * proizv
  if i == 2:
    return a2 * proizv
  if i == 3:
    return a3 * proizv
  if i == 4:
    return a4 * proizv

def p(proizv, i):
  if i == 1:
    return proizv_i(proizv, i) * b1 / k1
  if i == 2:
    return proizv_i(proizv, i) * b2 / k2
  if i == 3:
    return proizv_i(proizv, i) * b3 / k3
  if i == 4:
    return proizv_i(proizv, i) * b4 / k4

def R(proizv):
  return p(proizv, 1) + p(proizv, 2) + p(proizv, 3) +p(proizv, 4)

def u_i(proizv, i):
  if i == 1:
    return b1 / (1 - p(proizv, i))
  if i == 2:
    return b2 / (1 - p(proizv, i))
  if i == 3:
    return b3 / (1 - p(proizv, i))
  if i == 4:
    return b4 / (1 - p(proizv, i))
  
def U_(proizv):
  return u_i(proizv, 1) * a1 + u_i(proizv, 2) * a2 + u_i(proizv, 3) * a3 + u_i(proizv, 4) * a4

def w_i(proizv, i):
  if i == 1:
    return u_i(proizv, i) - b1
  if i == 2:
    return u_i(proizv, i) - b2
  if i == 3:
    return u_i(proizv, i) - b3
  if i == 4:
    return u_i(proizv, i) - b4

def W_(proizv):
  return w_i(proizv, 1) * a1 + w_i(proizv, 2) * a2 + w_i(proizv, 3) * a3 + w_i(proizv, 4) * a4

def l_i(proizv, i):
  return w_i(proizv, i) * proizv_i(proizv, i)

def L_(proizv):
  return l_i(proizv, 1) + l_i(proizv, 2) + l_i(proizv, 3) + l_i(proizv, 4)

def m_i(proizv, i):
  return u_i(proizv, i) * proizv_i(proizv, i)

def M_(proizv):
  return m_i(proizv, 1) + m_i(proizv, 2) + m_i(proizv, 3) + m_i(proizv, 4)
```

# Расчет характеристик для Марк (не экспонциальные)


```python
import math

b_3 = 0.5
v = 2
q = 2 / (1 + v ** 2)
q1 = q / 2
b_31 = (1 + math.sqrt((1-q1)/(2 * q1)*(v ** 2 - 1))) * b_3
b_32 = (1 - math.sqrt((q1)/(2 * (1 - q1))*(v ** 2 - 1))) * b_3
mu_31 = 1 / b_31
mu_32 = 1 / b_32 

print(f"q ≤ {q}")
print(f"Выбираем q = {q1}")
print(f"b'3(1) = {round(b_31, 4)} => μ'3(1) = {round(mu_31, 4)}")
print(f"b'3(2) = {round(b_32, 4)} => μ'3(2) = {round(mu_32, 4)}")
print(f"Проверка: q * b'31 + (1-q) * b'32 = {q1 * b_31 + (1 - q1) * b_32}")
```
